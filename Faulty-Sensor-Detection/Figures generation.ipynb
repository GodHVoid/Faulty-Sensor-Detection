```python
import pandas as pd
import numpy as np

data = pd.read_csv('data/result.csv',parse_dates=['time'], index_col=['time'])
data1 = pd.read_csv('New_data/LBNL_FDD_Dataset_SDAHU/AHU_annual.csv',)
sample = data.loc['2017-07-04 07:30:00':'2017-07-04 08:30:00']
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

plt.rcParams.update({'font.size': 32})
fig, ax = plt.subplots()
hours = mdates.HourLocator(interval = 1)
h_fmt = mdates.DateFormatter('%H:%M:%S')

ax.plot(sample.index, sample['zone_temp'].values, color = 'red', linewidth = 5, label = 'zone temp')
ax.plot(sample.index, sample['supply_temp'].values, color = 'green', linewidth = 5, label = 'supply temp')
ax.plot(sample.index, sample['occupancy'].values, color = 'black', linewidth = 5, label = 'occupancy')
ax.plot(sample.index, sample['outdoor_temp'].values, color = 'violet', linewidth = 5, label = 'outdoor temp')
#Then tick and format with matplotlib:
ax.xaxis.set_major_locator(hours)
ax.xaxis.set_major_formatter(h_fmt)

fig.autofmt_xdate()
fig.set_size_inches(18, 12)
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
data1 = pd.read_csv('New_data/LBNL_FDD_Dataset_SDAHU/AHU_annual.csv')


data1["Datetime"] = pd.to_datetime(data1["Datetime"])
print(data1)
data1 =data1[['Datetime','RA_TEMP','OA_TEMP', 'MA_TEMP','SA_TEMP','SA_CFM','ZONE_TEMP_1','ZONE_TEMP_2','ZONE_TEMP_3','ZONE_TEMP_4','ZONE_TEMP_5']]
data1['OA_TEMP'] = data1['OA_TEMP'] * 5/9 - 32 * 5/9 
data1['RA_TEMP'] = data1['RA_TEMP']* 5/9 - 32 * 5/9 
data1['MA_TEMP'] = data1['MA_TEMP']* 5/9 - 32 * 5/9 
data1['SA_TEMP'] = data1['SA_TEMP']* 5/9 - 32 * 5/9 
data1['ZONE_TEMP_1'] = data1['ZONE_TEMP_1']* 5/9 - 32 * 5/9 
data1['ZONE_TEMP_2'] = data1['ZONE_TEMP_2']* 5/9 - 32 * 5/9 
data1['ZONE_TEMP_3'] = data1['ZONE_TEMP_3']* 5/9 - 32 * 5/9 
data1['ZONE_TEMP_4'] = data1['ZONE_TEMP_4']* 5/9 - 32 * 5/9 
data1['ZONE_TEMP_5'] = data1['ZONE_TEMP_5']* 5/9 - 32 * 5/9 

data1['average_zone_temp'] = data1['ZONE_TEMP_1'] + data1['ZONE_TEMP_2'] +data1['ZONE_TEMP_3']+data1['ZONE_TEMP_4'] + data1['ZONE_TEMP_5']   
data1['average_zone_temp'] = data1['average_zone_temp']/5
#data1 = data1.loc["2018-01-01:01:00:00":"2018-02-01:01:00:00"]
```

```python
date = data1["Datetime"]
value = data1[["average_zone_temp", "OA_TEMP", "SA_TEMP"]]
```

```python
import matplotlib.dates as mdates
 # The "," is intentional.
fig, ax = plt.subplots(figsize=(40, 30))
monthly_locator = mdates.MonthLocator()
ax.xaxis.set_minor_locator(monthly_locator)
ax.plot(date, value['average_zone_temp'],linewidth = 5, label = 'zone temp')
ax.plot(date, value['OA_TEMP'],linewidth = 5, label = 'outside temp')
ax.plot(date, value['SA_TEMP'],linewidth = 5, label = 'supply temp')
plt.ylabel("C")
leg = plt.legend(loc='upper right')
fig.autofmt_xdate()
```
